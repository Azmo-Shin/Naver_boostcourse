 # melt로 데이터 전처리

```python
# melt로 지정한 칼럼을 기준으로 재정리
df_raw.melt(id_vars = ['국가(대륙)별', '상품군별', '판매유형별'])
```

```python
# melt를 활용해 기간에 입력 된 값들을 다시 재정리
df = df_raw.melt(id_vars = ['국가(대륙)별', '상품군별', '판매유형별'], var_name = '기간', value_name = '백만원')
print(df.shape)
print(df.head())
```

```python
df.head()
```

 # split으로 데이터 전처리

 ## 연도와 분기 만들기

```python
df['기간']
```

```python
'2020 3/4 p)'.split()
```

```python
'2020 3/4 p)'.split()[0]
```

```python
type('2020 3/4 p)'.split()[0])
```

```python
int('2020 3/4 p)'.split()[0])
```

```python
df['연도'] = df['기간'].map(lambda x : int(x.split()[0]))
df.head()
```

```python
'2020 3/4 p)'.split()[1].split('/')
```

```python
'2020 3/4 p)'.split()[1].split('/')[0]
```

```python
'2020 3/4 p)'.split()[1].split('/')[-1]
```

```python
df['분기'] = df['기간'].map(lambda x : int(x.split()[1].split('/')[0]))
df.head()
```

 ## 백만원 칼럼을 액수 표현

 * 2016년 이전 데이터는 결측치가 많고, '-'으로 기입됨. '-'로 표시된 값은 결측치이므로 NaN으로 대치(replace)

```python
df['백만원'].replace('-', np.nan)
```

```python
df['백만원'] = df['백만원'].replace('-', np.nan).astype(float)
df['백만원']
```

 # 결측값 제거

 * 값이 '합계'인 데이터를 삭제

```python
df = df[(df['국가(대륙)별'] != '합계') & (df['상품군별'] != '합계')].copy()
```

```python
# 결측값 확인
df.isnull().sum()
```

```python
# 판매유형별 칼럼은 각 도소매업으로 나누어져 입력이 되어 있으므로 결측값으로 잡히는 자료가 많으므로 제외
df_total = df[df['판매유형별'] == '계'].copy()
df_total.head()
```
